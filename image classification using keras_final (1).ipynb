```
#defining a schema of dataset
schema = StructType(df.select("image.*").schema.fields + [
    StructField("data_as_resized_array", ArrayType(IntegerType()), True),
    StructField("data_as_array", ArrayType(IntegerType()), True)
])
```

```
#function to decode an image array
def decode_img_array(img_array):
    return np.array(img_array).T


#function to Resize image
def resize_img(img_data, resize=True):
    mode = 'L' 
    img = Image.frombytes(mode=mode, data=img_data.data, size=[img_data.width, img_data.height])
    if (resize==True):
        img = img.resize([150,106],resample=Image.BICUBIC)

    arr = decode_img_array(np.asarray(img))
    
    if (resize==True):
        arr = arr.reshape([150*106])
    else:
        arr=arr.reshape([img_data.width*img_data.height])
 
    return arr

#function to create two extra columns in the dataframe keeping the original array as decoded and the resized image as decoded array
def resize_image_udf(dataframe_batch_iterator):
    
    for dataframe_batch in dataframe_batch_iterator:
    
        dataframe_batch["data_as_resized_array"] = dataframe_batch.apply(resize_img, args=(True,), axis=1)
        dataframe_batch["data_as_array"] = dataframe_batch.apply(resize_img, args=(False,), axis=1)
        yield dataframe_batch

resized_df1 = df.select('image.*').mapInPandas(resize_image_udf, schema)
resized_df2 = df2.select('image.*').mapInPandas(resize_image_udf, schema)
resized_df3 = df3.select('image.*').mapInPandas(resize_image_udf, schema)
resized_df4 = df4.select('image.*').mapInPandas(resize_image_udf, schema)
```

# Keras classification

```
#setting all the convolution layers for CNN
model = Sequential()
in_shape=(150,106, 1)
model.add(Conv2D(32, (3, 3), strides = (1, 1), padding='valid',name = 'conv0', input_shape = in_shape))
model.add(BatchNormalization(axis = 3, name = 'bn0'))
model.add(Activation('relu'))

model.add(Conv2D(32, (2,2), strides = (2,2), padding='valid',name="conv1"))

model.add(MaxPooling2D((3, 3),strides=(1,1),padding='valid', name='max_pool'))
model.add(Activation('relu'))

model.add(Conv2D(64, (2,2), strides = (2,2), padding='valid', name="conv2"))

model.add(AveragePooling2D((3, 3), name='avg_pool'))

model.add(GlobalAveragePooling2D())
model.add(Dense(300, activation="relu", name='rl'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid', name='sm'))
```

```
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop', 
              metrics=['accuracy'])
```

```
pand_df=final_train_df.toPandas()
```

```
pand_df.head(3)
```

```
X_tr=[]
imges_array = pand_df['data_as_resized_array'].values
for i in imges_array:
    i=i.reshape(150,106)
    temp = np.expand_dims(i, axis=2)
    X_tr.append(temp)    

X_tr = np.asarray(X_tr)
X_tr = X_tr.astype('float32')
#X_tr /= 255
Y_tr=pand_df['category'].values
Y_tr = np.asarray(Y_tr)
```

```
batch_size = 32
nb_epoch = 200
```

```
%%time
# Train model
history = model.fit(X_tr, Y_tr,
              batch_size=batch_size,
              epochs=nb_epoch,
              validation_split=0.2,
              shuffle=True,
              verbose=2)
```

```
history_df = pd.DataFrame(history.history)
history_df[['loss', 'val_loss']].plot()
```
